```python
# Import libraries
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Data dalam format array. Usia: Muda=0, Tua=1; Pendapatan: Rendah=0, Sedang=1, Tinggi=2; Status Pernikahan: Belum=0, Sudah=1; Membeli: Tidak=0, Ya=1
X = np.array([[0, 0, 0],  # Muda, Rendah, Belum
              [1, 2, 1],  # Tua, Tinggi, Sudah
              [1, 0, 1],  # Tua, Rendah, Sudah
              [0, 1, 0],  # Muda, Sedang, Belum
              [1, 1, 1]]) # Tua, Sedang, Sudah

y = np.array([0, 1, 1, 0, 1])  # Membeli: Tidak, Ya, Ya, Tidak, Ya

# Inisialisasi dan training model
clf = DecisionTreeClassifier(random_state=42, max_depth=3)
clf.fit(X, y)

# Plot decision tree
plt.figure(figsize=(10, 8))
plot_tree(clf, filled=True, rounded=True,
          feature_names=["Usia", "Pendapatan", "Status Pernikahan"],
          class_names=["Tidak", "Ya"])
plt.show()

# Prediksi untuk [Tua, Sedang, Belum]
prediction = clf.predict([[1, 1, 0]])
print("Prediksi untuk [Tua, Sedang, Belum]:", "Ya" if prediction[0] else "Tidak")
```

```python
# Importing libraries
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt


# Loading Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Creating the Decision Tree classifier and fitting the model
clf = DecisionTreeClassifier(criterion='entropy', random_state=0)
clf.fit(X, y)
# Plotting the Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()

```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_text
import graphviz
from sklearn.tree import export_graphviz
from IPython.display import display

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create Decision Tree model
clf = DecisionTreeClassifier(random_state=0)
clf.fit(X, y)

# Generate Graphviz data
dot_data = export_graphviz(
    clf, out_file=None,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True, rounded=True,
    special_characters=True
)

# Generate and display the graph
graph = graphviz.Source(dot_data)
display(graph)
```
